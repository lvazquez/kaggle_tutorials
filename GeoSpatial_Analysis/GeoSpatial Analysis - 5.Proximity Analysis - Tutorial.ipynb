## Measuring distance

To measure distances between points from two different **GeoDataFrames**, we first have to make sure that they use the same coordinate reference system (**CRS**). Thankfully, this is the case here, where both use **EPSG 2272**.

```python
print(stations.crs)
print(releases.crs)
```

We also need to check the **CRS** to see which units it uses (meters, feet, or something else). In this case, EPSG 2272 has units of feet and you can check this [here](https://epsg.io/2272).

Once these prerequisites are meet, it's relatively straightforward to compute distances in **GeoPandas**. The code cell below calculates the distance (in feet) between a relatively recent release incident and every station in the `stations` **GeoDataFrame**.


```python
# Select one release incident in particular
recent_release = releases.iloc[360]

# Measure distance from release to each station
distances = stations.geometry.distance(recent_release.geometry)
distances
```

Using the calculated distances we can obtain statistics, like the mean distance to each station or or the closest monitoring station to the incident:

```python
print('Mean distance to monitoring stations: {:.1f} feet\n'.format(distances.mean()))

print('Closest monitoring station ({:.1f} feet):'.format(distances.min()))
stations.iloc[distances.idxmin()][["ADDRESS", "LATITUDE", "LONGITUDE"]].to_frame().T
```

## Creating a buffer

If we want to understand all points on a map that are some radius away from a point, the simplest way is to create a buffer.

The code cell below creates a **GeoSeries** named `two_mile_buffer` containing 12 different Polygon objects. Each polygon is a buffer of 2 miles (or, 2*5280 feet) around a different air monitoring station.

```python
two_mile_buffer = stations.geometry.buffer(2*5280)
two_mile_buffer.head(3)
```
